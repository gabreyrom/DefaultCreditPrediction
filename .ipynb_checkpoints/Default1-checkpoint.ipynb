```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
default = pd.read_excel('default.xls')
```

```python
col_names = list(default.iloc[0,:])
```

```python
default.columns = col_names
default.drop('ID',inplace = True)
default.reset_index(drop = True, inplace = True)
```

```python
default.info()
```

```python
gender = np.array(default.SEX)
```

```python
gender.reshape(-1,1)
```

```python
encoder=OneHotEncoder()
gender_1hot=encoder.fit_transform(gender.reshape(-1,1))
```

```python
gender_1H_df = pd.DataFrame(gender_1hot.toarray(), columns = ['Male','Female'])
```

```python
default_encoded = pd.concat([default,gender_1H_df],axis=1)
```

```python
default_encoded.drop('SEX',axis = 1,inplace = True)
```

```python
education = np.array(default.EDUCATION)
```

```python
default_encoded.MARRIAGE.unique()
```

```python
encoder1=OneHotEncoder()
education_1hot=encoder1.fit_transform(education.reshape(-1,1))
```

```python
education_1H_df = pd.DataFrame(education_1hot.toarray(), columns = ['NA_E','Grad','Uni','HS','MS','Elem','NoE'])
#education_1H_df = pd.DataFrame(education_1hot.toarray(),colums = [''])
```

```python
default_encoded = pd.concat([default_encoded,education_1H_df],axis=1)
default_encoded.drop('EDUCATION',axis = 1,inplace = True)
```

```python
marriage = np.array(default.MARRIAGE)
```

```python
encoder2=OneHotEncoder()
marriage_1hot=encoder2.fit_transform(marriage.reshape(-1,1))
```

```python
marriage_1H_df = pd.DataFrame(marriage_1hot.toarray(), columns = ['NA_M','Married','Single','Other'])
```

```python
default_encoded = pd.concat([default_encoded,marriage_1H_df],axis=1)
default_encoded.drop('MARRIAGE',axis = 1,inplace = True)
```

```python
from sklearn import cross_validation
X_train, X_test, y_train, y_test =   cross_validation.train_test_split(default_encoded.drop(['default payment next month','LIMIT_BAL'],axis=1),   default_encoded['default payment next month'], test_size=0.10, random_state=111)
```

```python
default_encoded.drop(['default payment next month','LIMIT_BAL'],inplace=True,axis=1)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn import tree
from sklearn import svm
```

```python
model = GaussianNB()
#model = KNeighborsClassifier()
#model= RandomForestClassifier()
#model = tree.DecisionTreeClassifier(criterion='gini')
#model = svm.SVC()
```

```python
from sklearn.model_selection import train_test_split

# X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.33, random_state=42)

model.fit(list(X_train.drop('AGE',axis=1).values), list(y_train.values))

predicted = model.predict(list(X_test.drop('AGE',axis=1).values))
```

```python
from sklearn import metrics

#summarize the fit of the model
print(metrics.accuracy_score(list(y_test.values), predicted))

print(metrics.classification_report(list(y_test.values), list(y_test.values)))

print(metrics.confusion_matrix(list(y_test.values), list(y_test.values)))
```
